**Content based filtering on content matrix**

```python
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(content_data['content'])
book_content_matrix = pd.DataFrame(
    count_matrix.toarray(),
    index=content_data.index,
    columns=count.get_feature_names_out()
)

cosine_sim_content = cosine_similarity(book_content_matrix, book_content_matrix)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim_content, score_cutoff=70):
    user_input = title.lower()
    all_titles = list(indices.index)
    matches = process.extract(user_input, all_titles, limit=5, scorer=fuzz.partial_ratio)

    filtered = [(candidate, score) for (candidate, score) in matches if score >= score_cutoff]
    if not filtered:
        print(f"No similar titles (>= {score_cutoff} match).")
        return []

    best_match = filtered[0][0]
    best_score = filtered[0][1]

    book_id = indices[best_match]

    best_match_original_str = content_data.loc[book_id, 'original_title']
    print(f"Interpreting '{title}' as '{best_match_original_str}' (confidence: {best_score}).")
    print("Here are some recommendations:\n")

    row_idx = book_content_matrix.index.get_loc(book_id)

    sim_scores = list(enumerate(cosine_sim[row_idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:11]

    recommended_positions = [pos for (pos, sim) in sim_scores]

    recommended_book_ids = book_content_matrix.index[recommended_positions]

    recommended_books = content_data.loc[recommended_book_ids, 'original_title'].tolist()

    for rec in recommended_books:
        print(rec)
    return recommended_books
```

```python
def evaluate_content_mse(user_item_matrix, book_content_matrix):
    mse_scores = []

    for user_id in user_item_matrix.index:
        user_ratings = user_item_matrix.loc[user_id]

        rated_books = user_ratings[user_ratings != -1]

        if len(rated_books) < 2:
            continue

        train_books, test_books = train_test_split(rated_books.index, test_size=0.2, random_state=42)
        train_ratings = rated_books.loc[train_books]
        test_ratings = rated_books.loc[test_books]

        train_features = book_content_matrix.loc[train_books]
        user_profile = np.dot(train_ratings.values, train_features.values) / train_ratings.sum()

        test_features = book_content_matrix.loc[test_books]
        actual = test_ratings.values

        sim_scores = cosine_similarity([user_profile], test_features.values).flatten()

        predicted = (sim_scores + 1) / 2 * 5

        mse = mean_squared_error(actual, predicted)
        mse_scores.append(mse)

    overall_mse = np.mean(mse_scores) if mse_scores else float('nan')
    print(f"Average MSE for Content-Based approach: {overall_mse:.4f}")
    return overall_mse
```

```python
def get_user_books_loop():
    books_list = []
    while True:
        user_input = input("Enter a book title you like (or 'DONE' to finish): ").strip()

        if not user_input:
            print("No book title entered, stopping...\n")
            break

        if user_input.lower() == "done":
            print("Done entering books...\n")
            break

        books_list.append(user_input)

    return books_list
```

```python
def find_best_title(user_input, all_titles, cutoff=75):
    best_match = process.extractOne(
        user_input,
        all_titles,
        scorer=fuzz.partial_ratio
    )
    if best_match:
        title_candidate, score = best_match
        if score >= cutoff:
            return title_candidate
    return None
```

```python
def get_collaborative_recommendations(user_books_list):
    similar_books = np.zeros(book_corr.shape[0])

    for book in user_books_list:
        matched_title = find_best_title(book, book_titles, cutoff=0.75)
        if matched_title:
            print(f"Interpreting '{book}' as '{matched_title}'.")
            book_index = book_titles.index(matched_title)
            similar_books += book_corr[book_index]
        else:
            print(f"Warning: Book '{book}' not found in the dataset. Skipping.")

    book_preferences = [
        (book_titles[i], similar_books[i])
        for i in range(len(book_titles))
    ]

    book_preferences.sort(key=lambda x: x[1], reverse=True)
    return book_preferences
```

```python
def main():
    user_books = get_user_books_loop()
    if not user_books:
        print("No books entered. Exiting.")
        return

    all_recommendations = get_collaborative_recommendations(user_books)

    print("\nTop 9 recommendations (excluding books you already like):\n")
    i = 0
    n = 0
    while n < 9 and i < len(all_recommendations):
        candidate_book = all_recommendations[i][0]
        i += 1

        if candidate_book in user_books:
            continue

        print(candidate_book)
        n += 1
```

```python
def compute_item_item_mse(book_rating, book_corr):
    user_mse = []

    for user_id in book_rating.index:
        user_row = book_rating.loc[user_id]
        rated_books = user_row[user_row != 0]
        if len(rated_books) < 2:
            continue

        train_books, test_books = train_test_split(rated_books.index, test_size=0.2, random_state=42)
        train_ratings = rated_books.loc[train_books]
        test_ratings = rated_books.loc[test_books]

        preds = []
        actuals = []

        for test_book in test_books:
            try:
                test_book_idx = book_rating.columns.get_loc(test_book)
            except KeyError:
                continue

            num = 0.0
            den = 0.0
            for tb in train_books:
                tb_idx = book_rating.columns.get_loc(tb)
                sim = book_corr[test_book_idx, tb_idx]
                num += sim * train_ratings[tb]
                den += abs(sim)

            if den == 0:
                pred_rating = train_ratings.mean()
            else:
                pred_rating = num / den

            preds.append(pred_rating)
            actuals.append(test_ratings[test_book])

        if preds:
            user_mse.append(mean_squared_error(actuals, preds))

    overall_mse = np.mean(user_mse) if user_mse else float('nan')
    return overall_mse

mse_val = compute_item_item_mse(book_rating, book_corr)
print("Item–Item CF MSE:", mse_val)
```

```python
new_book_titles = new_book_rating.columns.tolist()
book_to_idx = {book: idx for idx, book in enumerate(new_book_titles)}
idx_to_book = {idx: book for book, idx in book_to_idx.items()}
```

```python
book_rating_matrix = new_book_rating.T
book_rating_sparse = csr_matrix(book_rating_matrix.values)
print("Converted to sparse book-rating matrix.")
```

```python
n_components = 100
svd = TruncatedSVD(n_components=n_components, random_state=42)
book_rating_reduced = svd.fit_transform(book_rating_sparse)
print(f"Applied TruncatedSVD with {n_components} components.")
print(f"Reduced book-rating matrix shape: {book_rating_reduced.shape}")
```

```python
knn = NearestNeighbors(metric='cosine', algorithm='brute', n_jobs=-1)
knn.fit(book_rating_reduced)
print("Fitted kNN on reduced data.")
```

```python
def get_user_books_loop_new():
    new_books_list = []
    while True:
        user_input = input("Enter a book title you like (or 'DONE' to finish): ").strip()

        if not user_input:
            print("No book title entered, stopping...\n")
            break

        if user_input.lower() == "done":
            print("Done entering books...\n")
            break

        new_books_list.append(user_input)

    return new_books_list
```

```python
def knn_recommend(user_books, k=20):
    if not user_books:
        print("No books were provided. Exiting...")
        return

    similar_books_scores = np.zeros(len(new_book_titles))

    print("\nInterpreting your input:")
    for book in user_books:
        if not isinstance(book, str):
            print(f"Invalid book entry: {book}. Skipping...")
            continue

        best_match = process.extractOne(book, new_book_titles, scorer=fuzz.partial_ratio)

        if not best_match or best_match[1] < 75:
            print(f"'{book}' not found in the dataset. Skipping...")
            continue

        matched_book = best_match[0]
        print(f"- '{book}' interpreted as '{matched_book}'.")

        book_index = book_to_idx.get(matched_book)
        if book_index is None:
            print(f"'{matched_book}' not found in book indices. Skipping...")
            continue

        if book_index >= book_rating_reduced.shape[0]:
            print(f"Index {book_index} for book '{matched_book}' is out of bounds. Skipping...")
            continue

        book_vector = book_rating_reduced[book_index].reshape(1, -1)

        distances, indices = knn.kneighbors(book_vector, n_neighbors=k+1)
        for i, idx in enumerate(indices.flatten()[1:]):
            similar_books_scores[idx] += 1 - distances.flatten()[i + 1]

    sorted_books = sorted(
        [(idx_to_book[i], score) for i, score in enumerate(similar_books_scores)],
        key=lambda x: x[1],
        reverse=True
    )

    recommendations = [book for book, score in sorted_books if book not in user_books][:k]

    if recommendations:
        print("\nRecommended books:")
        for rec in recommendations:
            print(f"- {rec}")
        print("These are my recommended books, I hope you like them :)")
    else:
        print("\nNo recommendations could be generated based on the input.")
```

```python
def main_2():
    print("Welcome to the KNN Book Recommender!")
    print("Enter the titles of books you like. When you're done, type 'DONE'.")

    user_books = get_user_books_loop_new()

    if not user_books:
        print("No books were entered. Exiting...")
        return

    print("\nProcessing your recommendations...")
    knn_recommend(user_books, k=20)
```

```python
def evaluate_knn_item_item_mse(book_rating, k=5, n_jobs=-1):
    book_rating_sparse_eval = csr_matrix(book_rating_matrix.values)
    print("Converted book_rating to sparse matrix for evaluation.")

    book_rating_reduced_eval = svd.transform(book_rating_sparse_eval)
    print("Applied TruncatedSVD transformation on evaluation data.")
    print(f"Reduced book-rating evaluation matrix shape: {book_rating_reduced_eval.shape}")

    user_ids = book_rating.index.tolist()
    book_titles = new_book_titles

    def predict_rating(test_book, train_books, train_ratings):
        test_book_idx = book_to_idx.get(test_book)
        if test_book_idx is None:
            return None

        test_vector = book_rating_reduced_eval[test_book_idx].reshape(1, -1)

        distances, indices = knn.kneighbors(test_vector, n_neighbors=k+1)
        distances = distances.flatten()[1:]
        indices = indices.flatten()[1:]

        num = 0.0
        den = 0.0

        for dist, neighbor_idx in zip(distances, indices):
            neighbor_book = idx_to_book.get(neighbor_idx)
            if neighbor_book in train_books:
                sim = 1.0 - dist
                user_rating_for_neighbor = train_ratings[neighbor_book]
                num += sim * user_rating_for_neighbor
                den += abs(sim)

        if den == 0:
            pred_rating = train_ratings.mean()
        else:
            pred_rating = num / den

        return pred_rating

    def process_user(user_id):
        user_row = book_rating.loc[user_id]
        rated_books = user_row[user_row != 0]
        if len(rated_books) < 2:
            return None

        train_books, test_books = train_test_split(rated_books.index, test_size=0.2, random_state=42)
        train_ratings = rated_books.loc[train_books]
        test_ratings = rated_books.loc[test_books]

        preds = []
        actuals = []

        for test_book in test_books:
            pred = predict_rating(test_book, train_books, train_ratings)
            actual = test_ratings[test_book]

            if pred is not None:
                preds.append(pred)
                actuals.append(actual)

        if preds:
            return mean_squared_error(actuals, preds)
        return None

    user_mses = Parallel(n_jobs=n_jobs)(
        delayed(process_user)(user_id) for user_id in user_ids
    )

    user_mses = [mse for mse in user_mses if mse is not None]
    overall_mse = np.mean(user_mses) if user_mses else float('nan')
    print(f"KNN Item-Item MSE: {overall_mse}")
    return overall_mse
```

```python
def recommend_books_cnn(user_title, tokenizer, embedding_model, book_embeddings, new_content_data, top_n=10, threshold=0.85):
    user_sequence = tokenizer.texts_to_sequences([user_title])
    user_sequence = pad_sequences(user_sequence, maxlen=80, padding='post', truncating='post')

    user_embedding = embedding_model.predict(user_sequence)

    similarities = cosine_similarity(user_embedding, book_embeddings).flatten()

    filtered_indices = [i for i, score in enumerate(similarities) if score >= threshold]
    filtered_indices = sorted(filtered_indices, key=lambda i: similarities[i], reverse=True)[:top_n]

    recommended_books = [
        (new_content_data.iloc[i]['Title'], similarities[i])
        for i in filtered_indices
    ]

    if not recommended_books:
        print("No recommendations found with the given threshold.")
        return []

    print("\nTop Recommendations:")
    for book, score in recommended_books:
        print(f"- {book} (Similarity: {score:.4f})")

    return recommended_books
```
